```python
# Q2. Consider the Principal component analysis method
# a. What are the hyperparameters of the PCA?
# # Ans 2.a
# Hyperparameters are important because they directly control the behaviour of the given algorithm and have a 
# significant impact on the performance of algorithm.
# One of the hyperparamters of PCA is k or n_components (in sklearn) or principal components, this k value tells the algorithm to which dimension
# we have to project the dimension space of original data. 
# This is to note that after dimensionality reduction the components hold no meaning to it.
# For example, in our original dataset we have 4 dimensions and we set the value of k to 2, then the PCA algorithm will
# project the original 4 dimensional space to that of 2.
```

```python
# b. Show how these hyperparameters affect dimensionality reduction of your data? (by plot/graph/any other means)
# Ans 2.b

# Importing the libraries
import numpy as np
import os
from PIL import Image
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Loading the dataset
w1 = np.array([[0.42, -0.2, 1.3, 0.39, -1.6, -0.029, -0.23, 0.27, -1.9, 0.87, -0.4, -0.31, 0.38, -0.15, -0.35, 0.17, -0.011, -0.27, -0.065, -0.12, 0.83, 1.1, -0.44, 0.047, 0.28, -0.39, 0.34, -0.3, 1.1, 0.18], [-0.087, -3.3, -0.32, 0.71, -5.3, 0.89, 1.9, -0.3, 0.76, -1.0, 0.58, 0.27, 0.055, 0.53, 0.47, 0.69, 0.55, 0.61, 0.49, 0.054, 1.6, 1.6, -0.41, -0.45, 0.35, -0.48, -0.079, -0.22, 1.2, -0.11], [0.58, -3.4, 1.7, 0.23, -0.15, -4.7, 2.2, -0.87, -2.1, -2.6, 0.089, -0.04, -0.035, 0.011, 0.034, 0.1, -0.18, 0.12, 0.0012, -0.063, -0.014, 0.48, 0.32, 1.4, 3.1, 0.11, 0.14, 2.2, -0.46, -0.49]])
print(w1)
```

```python
# Projection of datapoints in 3D
fig = plt.figure()
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(w1[0], w1[1], w1[2], s=30, marker='s', c='grey', edgecolor='black')
```

```python
w1 = w1.T
```

```python
print(w1)
```

```python
# Calculating the mean 
mean_w1 = np.mean(w1.T, axis=1)
print(mean_w1)
```

```python
# Calculating the scatter matrix
c_w1 = w1 - mean_w1
c_w1 = np.asarray(c_w1)
print(c_w1)
```

```python
# Calculating the covariance matrix
v_w1 = np.cov(c_w1.T)
print(v_w1)
```

```python
# Calculating the eigen vector and eigen values
eigval, eigvec = np.linalg.eig(v_w1)
print(eigvec)
```

```python
# Choosing k-different eigen vectors
eigen_pairs = [[np.abs(eigval[i]),eigvec[:,i]] for i in range(len(eigval))]
eigen_pairs = sorted(eigen_pairs,key=lambda k: k[0],reverse=True)
w0 = eigvec
print(w0)
w1 = np.hstack((eigen_pairs[0][1][:,np.newaxis].real, eigen_pairs[1][1][:,np.newaxis].real))
print(w1)
w2 = np.hstack((eigen_pairs[0][1][:,np.newaxis].real))
print(w2)
```

```python
print(c_w1)
```

```python
# Applying the k choosen eigen vectors to original dataset
y0 = w0.T.dot(c_w1.T)
y1 = w1.T.dot(c_w1.T)
y2 = w2.T.dot(c_w1.T)
print(y0)
print(y1)
print(y2)
```

```python
# Projection of datapoints when k=3 with discriminant line
fig0 = plt.figure()
fig0 = plt.figure(figsize=(10,10))
ax0 = fig0.add_subplot(111, projection='3d')
ax0.scatter(y0.T[:,0], y0.T[:,1], y0.T[:,2], s=30, marker='s', c='grey', edgecolor='black')
ax0.plot(w0[:,0], w0[:,1], w0[:,2])
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('Fig 1. K = 3')
```

```python
# Projection of datapoints when k=2 with discriminant line
ax1 = plt.subplot(111)
plt.scatter(y1.T[:,0], y1.T[:,1], marker='s', c='grey', edgecolor='black')
plt.plot(w1[:,0], w1[:,1])
plt.xlabel('PC1')
plt.ylabel('PC2')
leg = plt.legend(loc='upper right', fancybox=True)
leg.get_frame().set_alpha(0.5)
plt.tick_params(axis="both", which="both", bottom="off", top="off", labelbottom="on", left="off", right="off", labelleft="on")
ax.spines["top"].set_visible(False)  
ax.spines["right"].set_visible(False)
ax.spines["bottom"].set_visible(False)
ax.spines["left"].set_visible(False)
plt.title('Fig 2. K = 2')
```

```python
# Projection of datapoints when k=1 with discriminant line
ax2 = plt.subplot(111)
plt.scatter(y2.T, np.zeros_like(y2.T) + 0.5, marker='s', c='grey', edgecolor='black')
plt.plot(w2)
plt.xlabel('PC1')
plt.ylabel('PC2')
leg = plt.legend(loc='upper right', fancybox=True)
leg.get_frame().set_alpha(0.5)
plt.tick_params(axis="both", which="both", bottom="off", top="off", labelbottom="on", left="off", right="off", labelleft="on")
ax.spines["top"].set_visible(False)  
ax.spines["right"].set_visible(False)
ax.spines["bottom"].set_visible(False)
ax.spines["left"].set_visible(False)
plt.title('Fig 3. K = 1')
```
