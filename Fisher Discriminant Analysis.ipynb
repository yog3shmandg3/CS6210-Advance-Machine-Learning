```python
# Q1. Consider the fisher linear discriminant method.
# a. Write a general program to calculate the optimal direction w for a Fisher linear discriminant based on the three-dimensional data.
# Ans 1.a

# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Loading the dataset
w1 = np.array([[0.42, -0.2, 1.3, 0.39, -1.6, -0.029, -0.23, 0.27, -1.9, 0.87], [-0.087, -3.3, -0.32, 0.71, -5.3, 0.89, 1.9, -0.3, 0.76, -1.0], [0.58, -3.4, 1.7, 0.23, -0.15, -4.7, 2.2, -0.87, -2.1, -2.6]])
w2 = np.array([[-0.4, -0.31, 0.38, -0.15, -0.35, 0.17, -0.011, -0.27, -0.065, -0.12], [0.58, 0.27, 0.055, 0.53, 0.47, 0.69, 0.55, 0.61, 0.49, 0.054], [0.089, -0.04, -0.035, 0.011, 0.034, 0.1, -0.18, 0.12, 0.0012, -0.063]])
w3 = np.array([[0.83, 1.1, -0.44, 0.047, 0.28, -0.39, 0.34, -0.3, 1.1, 0.18], [1.6, 1.6, -0.41, -0.45, 0.35, -0.48, -0.079, -0.22, 1.2, -0.11], [-0.014, 0.48, 0.32, 1.4, 3.1, 0.11, 0.14, 2.2, -0.46, -0.49]])
print(w1)
print(w2)
print(w3)
```

```python
# Projection of datapoints of all the three classes in 3D
fig = plt.figure()
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(w1[0], w1[1], w1[2], s=30, marker='s', c='grey', edgecolor='black')
ax.scatter(w2[0], w2[1], w2[2], s=30, marker='^', c='yellow', edgecolor='black')
ax.scatter(w3[0], w3[1], w3[2], s=30, marker='o', c='blue', edgecolor='black')
```

```python
# Calculating mean for each class
mean_w1 = np.mean(w1, axis=1).reshape(3,1)
mean_w2 = np.mean(w2, axis=1).reshape(3,1)
mean_w3 = np.mean(w3, axis=1).reshape(3,1)
print(mean_w1)
print(mean_w2)
print(mean_w3)
```

```python
# Calculating the scatter matrix and scatter between the classes
scatter_w1 = np.dot((w1 - mean_w1), (w1 - mean_w1).T)
scatter_w2 = np.dot((w2 - mean_w2), (w2 - mean_w2).T)
scatter_w3 = np.dot((w3 - mean_w3), (w3 - mean_w3).T)
print(scatter_w1)
print(scatter_w2)
print(scatter_w3)
sw = scatter_w1 + scatter_w2 + scatter_w3
print(sw)
```

```python
# Calculating the overall mean
# mean_vec = []
# mean_vec.append(mean_w1)
# mean_vec.append(mean_w2)
# mean_vec.append(mean_w3)
# mean_vec = np.asarray(mean_vec)
mean_vec = np.hstack([mean_w1, mean_w2, mean_w3])
print(mean_vec)
ov_mean = np.mean(mean_vec, axis=1).reshape(3,1)
print(ov_mean)
```

```python
# Calculating the scatter between the classes
n = 10
sb_w1 = n * np.dot((mean_w1 - ov_mean), (mean_w1 - ov_mean).T)
sb_w2 = n * np.dot((mean_w2 - ov_mean), (mean_w2 - ov_mean).T)
sb_w3 = n * np.dot((mean_w3 - ov_mean), (mean_w3 - ov_mean).T)
print(sb_w1)
print(sb_w2)
print(sb_w3)
sb = sb_w1 + sb_w2 + sb_w3
print(sb)
```

```python
# Calculating the eigervector and eigenvalues
eigval, eigvec = np.linalg.eig(np.dot(np.linalg.inv(sw), sb))
print(eigval)
print(eigvec)
```

```python
# Choosing the best two eigen pair that discriminate best between the dataset
eigen_pairs = [[np.abs(eigval[i]),eigvec[:,i]] for i in range(len(eigval))]
eigen_pairs = sorted(eigen_pairs,key=lambda k: k[0],reverse=True)
w = np.hstack((eigen_pairs[0][1][:,np.newaxis].real,eigen_pairs[1][1][:,np.newaxis].real))
print(w)
```

```python
# Applying the vector to the original class w1
w_1 = w1.transpose()
y1 = w_1.dot(w)
print(y1)
```

```python
# Applying the vector to the original class w2
y2 = w2.T.dot(w)
print(y2)
```

```python
# Applying the vector to the original class w3
y3 = w3.T.dot(w)
print(y3)
```

```python
print(y1[:,0])
print(y1[:,1])
```

```python
# Projection of datapoints of all the three classes in 2D with discriminant line
ax = plt.subplot(111)
plt.scatter(y1[:,0], y1[:,1], marker='s', c='grey', edgecolor='black')
plt.scatter(y2[:,0], y2[:,1], marker='^', c='yellow', edgecolor='black')
plt.scatter(y3[:,0], y3[:,1], marker='o', c='green', edgecolor='black')
plt.plot(w)
plt.xlabel('FD1')
plt.ylabel('FD2')
leg = plt.legend(loc='upper right', fancybox=True)
leg.get_frame().set_alpha(0.5)
plt.tick_params(axis="both", which="both", bottom="off", top="off", labelbottom="on", left="off", right="off", labelleft="on")
ax.spines["top"].set_visible(False)  
ax.spines["right"].set_visible(False)
ax.spines["bottom"].set_visible(False)
ax.spines["left"].set_visible(False)
```

```python
# b. Find the optimal w for categories w2 and w3 in the table above.
# Ans 1.b
# Projection of datapoints of classes w2 and w3 in 3D
fig2 = plt.figure()
fig2 = plt.figure(figsize=(10, 10))
ax2 = fig2.add_subplot(111, projection='3d')
ax2.scatter(w2[0], w2[1], w2[2], s=30, marker='^', c='yellow', edgecolor='black')
ax2.scatter(w3[0], w3[1], w3[2], s=30, marker='o', c='blue', edgecolor='black')
```

```python
# Calculating mean for both class
b_mean_w2 = np.mean(w2, axis=1).reshape(3,1)
b_mean_w3 = np.mean(w3, axis=1).reshape(3,1)
print(b_mean_w2)
print(b_mean_w3)
```

```python
# Calculating the scatter matrix and scatter between the classes
b_scatter_w2 = np.dot((w2 - b_mean_w2), (w2 - b_mean_w2).T)
b_scatter_w3 = np.dot((w3 - b_mean_w3), (w3 - b_mean_w3).T)
print(b_scatter_w2)
print(b_scatter_w3)
b_sw = b_scatter_w2 + b_scatter_w3
print(b_sw)
```

```python
# Calculating the overall mean
b_mean_vec = np.hstack([b_mean_w2, b_mean_w3])
print(b_mean_vec)
b_ov_mean = np.mean(b_mean_vec, axis=1).reshape(3,1)
print(b_ov_mean)
```

```python
# Calculating the scatter between the classes
n = 10
b_sb_w2 = n * np.dot((b_mean_w2 - b_ov_mean), (b_mean_w2 - b_ov_mean).T)
b_sb_w3 = n * np.dot((b_mean_w3 - b_ov_mean), (b_mean_w3 - b_ov_mean).T)
print(b_sb_w2)
print(b_sb_w3)
b_sb = b_sb_w2 + b_sb_w3
print(b_sb)
```

```python
# Calculating the eigervector and eigenvalues
b_eigval, b_eigvec = np.linalg.eig(np.dot(np.linalg.inv(b_sw), b_sb))
print(b_eigval)
print(b_eigvec)
```

```python
# Choosing the best two eigen pair that discriminate best between the dataset
b_eigen_pairs = [[np.abs(b_eigval[i]),b_eigvec[:,i]] for i in range(len(b_eigval))]
b_eigen_pairs = sorted(b_eigen_pairs,key=lambda k: k[0],reverse=True)
b_w = np.hstack((b_eigen_pairs[0][1][:,np.newaxis].real,b_eigen_pairs[1][1][:,np.newaxis].real))
print(b_w)
```

```python
# Applying the vector to the original class w2
b_y2 = w2.T.dot(b_w)
print(b_y2)
```

```python
# Applying the vector to the original class w3
b_y3 = w3.T.dot(b_w)
print(b_y3)
```

```python
# c. Plot a line representing your optimal direction w and mark on it the positions of the projected points
# Ans 1.c
# Projection of datapoints of all the three classes in 2D with discriminant line
ax2 = plt.subplot(111)
plt.scatter(b_y2[:,0], b_y2[:,1], marker='^', c='yellow', edgecolor='black')
plt.scatter(b_y3[:,0], b_y3[:,1], marker='o', c='green', edgecolor='black')
plt.plot(b_w[:,0],b_w[:,1])

plt.xlabel('FD1')
plt.ylabel('FD2')
leg = plt.legend(loc='upper right', fancybox=True)
leg.get_frame().set_alpha(0.5)
plt.tick_params(axis="both", which="both", bottom="off", top="off", labelbottom="on", left="off", right="off", labelleft="on")
ax.spines["top"].set_visible(False)  
ax.spines["right"].set_visible(False)
ax.spines["bottom"].set_visible(False)
ax.spines["left"].set_visible(False)
```
